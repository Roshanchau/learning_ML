Intro to matplotlib

```python
import matplotlib.pyplot as plt
```

LINE PLOT

```python
month=['jan' , 'feb' , 'mar' , 'who' , 'yanesh']
growth=[3,2 ,5,5.5,3]
```

```python
plt.plot(month, growth , 'o-')
```

```python
plt.plot(month, growth , 'o-' , color='red')
```

```python
plt.plot(month, growth , 'o--' , color='red')
plt.xlabel("month")
plt.ylabel("Growth in (%)")
plt.title("growth by month")
plt.show()
```

Barplot

```python
import seaborn as sms
df=sms.load_dataset('iris')
df.head(3)
```

```python
df.species.unique()
```

```python
t=df.species.value_counts()
t
```

```python
plt.bar(x=t.index, height=t.values , color=["brown", "red" , "pink"])
plt.xlabel("species")
plt.ylabel("frequency")
```

```python
sms.countplot(x='sepal_length' ,data=df) #barplot of a column
```

PIEEEEEEE CHARTTTTTTTTTTTTTTTTTTTTTTTT

```python
s=df.species.value_counts()
plt.pie(x=s.values, labels=s.index, autopct="%.1f%%")
```
