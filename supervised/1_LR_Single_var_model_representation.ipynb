```python
w=100
b=100
print(f"w: {w}")
print(f"b: {b}")
```

```python
def compute_model_output(x, w, b):
    """
    Computes the prediction of a linear model
    Args:
      x (ndarray (m,)): Data, m examples 
      w,b (scalar)    : model parameters  
    Returns
      f_wb (ndarray (m,)): model prediction
    """
    m=x.shape[0]
    f_wb=np.zeros(m) #creates a 1d numpy array with m entries.
    for i in range(m):
        f_wb[i]=x[i]*w+b
    return f_wb
```

let's compute the compute_model_output function and plot the output

```python
hehe_f_wb=compute_model_output(x_train, w,b)

# plot our model prediction
plt.plot(x_train, hehe_f_wb, c='b', label='our prediction')

# plot the data points
plt.scatter(x_train, y_train, marker='x' , c='r' , label='Actual values')

# set the title
plt.title('Housing Prices')

# set the x_label
plt.xlabel('size (in 1000sq.ft)')

# set the y_label
plt.ylabel('prices (in 1000$)')

plt.legend()
plt.show()
```
