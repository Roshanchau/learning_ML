The sigmoid function is implemented in python as shown in the cell below.

```python
def sigmoid(z):
    g=1/(1+np.exp(-z))
    return g
```

let's see what the output of this function if for various value of z

```python
# generate an array of evenly spaced values between -10 and 10
z_tmp=np.arange(-10, 11)

# use the function implemented above to get the sigmoid function
y=sigmoid(z_tmp)

# code for pretty printing the two arrays_next to each other
np.set_printoptions(precision=3)
print("Input (z) , Output (sigmoid(z))")
print(np.c_[z_tmp , y])
```

```python
# plot z vs sgmoid(z)
fig,ax=plt.subplots(1,1 ,figsize=(5,3))
ax.plot(z_tmp , y , c="b")

ax.set_title("sigmoid function")
ax.set_ylabel('sigmoid(z)')
ax.set_xlabel('z')
draw_vthresh(ax, 0) #ax means plot garne thau.
```
